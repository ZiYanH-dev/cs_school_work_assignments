```python
import numpy as np

# 模拟训练集标签（一维，N=5）
y_train = np.array([10, 20, 30, 40, 50])  # shape: (5,)

# 模拟k_indices（二维，M=2个测试样本，k=3个近邻）
k_indices = np.array([
    [0, 2, 4],  # 第1个测试样本的3个近邻索引：0、2、4
    [1, 3, 0]   # 第2个测试样本的3个近邻索引：1、3、0
])  # shape: (2, 3)

# 用二维索引取一维数组的值
k_neighbor_labels = y_train[k_indices]
print(k_neighbor_labels)
# 输出：
# [[10 30 50]
#  [20 40 10]]
print(k_neighbor_labels.shape)  # 输出：(2, 3) → 和k_indices形状一致！

print(np.bincount(y_train))
```
