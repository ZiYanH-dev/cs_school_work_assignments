## Background: Connect Four

Connect Four is a two-player game played on a vertical grid:
1. Players take turns dropping tokens into one of the columns, and each token falls to the lowest available position in that column due to gravity. 

2. The goal is to be the first player to connect four of their own tokens in a row, either horizontally, vertically, or diagonally.

3. Unlike Gomoku, players cannot place a token on any empty cell. Because of gravity, moves are restricted to the lowest empty position of a column, which makes the state space smaller and more suitable for minimax-based search.
